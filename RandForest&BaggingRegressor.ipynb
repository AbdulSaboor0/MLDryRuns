```python
df = df.drop(['car_name', 'brand'], axis = 1)
```

```python
df.head()
```

```python
df.describe()
```

```python
X = df.drop('selling_price', axis =1)
y = df['selling_price']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['model'] = le.fit_transform(X['model'])
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
num_features = X.select_dtypes(exclude= 'object').columns
Oh_columns = ['seller_type', 'fuel_type', 'transmission_type']
```

```python
num_transform = StandardScaler()
oh_transform = OneHotEncoder(drop ='first')

preprocessor = ColumnTransformer([
    ('OneHotEncoder', oh_transform, Oh_columns),
    ('StandardScaler', num_transform, num_features )
], remainder = 'passthrough')
```

```python
X = preprocessor.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor, BaggingRegressor

bag_Reg = BaggingRegressor()
randomF = RandomForestRegressor()
```

```python
bag_Reg.fit(X_train,y_train)
```

```python
pred = bag_Reg.predict(X_test)
```

```python
bag_Reg.oob_score
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(pred, y_test)
r2
```

```python
randomF.fit(X_train,y_train)
```

```python
pred = randomF.predict(X_test)
from sklearn.metrics import r2_score
r2 = r2_score(pred, y_test)
r2
```
