```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples = 1000, n_features = 2 , centers = 3)
```

```python
X1 = X
X = pd.DataFrame(X)
X.columns = ['Column1','Column2']
X1
```

```python
sns.scatterplot(x = X['Column1'], y = X['Column2'], data = X)
```

```python
plt.scatter(X1[:,0], X1[:,1], c = y)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test, y_train,y_test = train_test_split(X,y , test_size = 0.3)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for k in range(1,11):
  cluster = KMeans(n_clusters = k)
  cluster.fit(X_train_scaled)
  wcss.append(cluster.inertia_)

```

```python
wcss
```

### ***Plotting the Elbow Curve***

```python
plt.plot(range(1,11), wcss)
plt.xticks(range(1,11))
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
cluster = KMeans(n_clusters = 2)
```

```python
cluster.fit_predict(X_train_scaled)
```

```python
pred = cluster.predict(X_test_scaled)
```

### ***Silhoutte Score***

```python
from sklearn.metrics import silhouette_score
sil_coeff =[]
for k in range (2,11):
  means = KMeans(n_clusters= k)
  means.fit(X_train_scaled)
  score = silhouette_score(X_train_scaled, means.labels_)
  sil_coeff.append(score)
```

```python
sil_coeff
```

```python
plt.plot(range(2,11), sil_coeff)
plt.xticks(range(2,11))
plt.xlabel('No. of clusters')
plt.ylabel('Silhoutte Coefficients')
plt.show()
```
