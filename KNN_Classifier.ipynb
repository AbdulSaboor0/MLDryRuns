```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification

X,y = make_classification(n_samples = 1000, n_features = 4 , n_redundant =1,n_classes= 2,n_clusters_per_class = 2, n_informative= 2)
```

```python
X = pd.DataFrame(X)
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test = train_test_split(X,y, test_size = 0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier

Classifier = KNeighborsClassifier()
```

```python
Classifier.fit(X_train,y_train)
```

```python
pred = Classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

rep = classification_report(pred, y_test)
cm = confusion_matrix(pred, y_test)
acc = accuracy_score(pred, y_test)
```

```python
print(rep)
print(cm)
print(acc)
```

```python
params = { 'n_neighbors': [6,7,9,10,4,3,12],
          'algorithm': [ 'ball_tree', 'kd_tree', 'brute'],
          'p': [1,2] ,
          'leaf_size' : [28,26,30,34,36,38,40]
          }
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator = Classifier, param_grid = params, cv = 5, verbose =3
                    )
```

```python
grid.fit(X_train, y_train)
```

```python
pred = grid.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

rep = classification_report(pred, y_test)
cm = confusion_matrix(pred, y_test)
acc = accuracy_score(pred, y_test)
```

```python
print(rep)
print(cm)
print(acc)
```
